## Regression

## LinearRegression | BayesianRidge | DecisionTree | Support Vector Machines

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression, BayesianRidge  # Linear Regression and Bayesian Linear Regression
from sklearn.ensemble import RandomForestRegressor  # Regression Trees
from sklearn.tree import DecisionTreeRegressor  # Regression Trees
from sklearn.svm import SVR  # Non-Linear Regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline  # Polynomial Regression
```

```python
# Load the datasets
train_data = pd.read_csv('C:/Zintern/python/Algoritms/Regression/mobile/train.csv')
test_data = pd.read_csv('C:/Zintern/python/Algoritms/Regression/mobile/test.csv')
```

```python
# Preprocess the data
def preprocess_data(df):
    print(df.isnull().sum())
    return df
```

```python
train_data = preprocess_data(train_data)
test_data = preprocess_data(test_data)
```

```python
# Split the data into features and target labels
X_train = train_data.drop(columns=['Price'])
y_train = train_data['Price']
X_test = test_data
```

```python
# Initialize the models
models = {
    'Linear Regression': LinearRegression(),  # Linear Regression
    'Decision Tree': DecisionTreeRegressor(),  # Regression Trees
    'Random Forest': RandomForestRegressor(),  # Regression Trees
    'Support Vector Machine': SVR(),  # Non-Linear Regression
    'Polynomial Regression': make_pipeline(PolynomialFeatures(degree=2), RandomForestRegressor())  # Polynomial Regression
}
```

```python
# Evaluate each model
accuracies = {}
for name, model in models.items():
    # Use R² scoring for regression evaluation
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
    accuracies[name] = scores.mean()
```

```python
# Print accuracies for comparison
print("Model Accuracies:")
for name, score in accuracies.items():
    print(f'{name} Accuracy: {score:.4f}')
```

```python
# Identify the model with the highest accuracy
best_model_name = max(accuracies, key=accuracies.get)
best_model = models[best_model_name]
```

```python
print(f'Best Model: {best_model_name} with accuracy {accuracies[best_model_name]:.4f}')
```

```python
# Perform cross-validation for the best model
cross_val_scores = cross_val_score(best_model, X_train, y_train, cv=5, scoring='r2')
print(f'Cross-Validated Accuracy for {best_model_name}: {cross_val_scores.mean():.4f} ± {cross_val_scores.std():.4f}')
```

```python
# Train the best model on the entire training set
best_model.fit(X_train, y_train)
```

```python
# Predict on the test set
test_predictions = best_model.predict(X_test)
```

```python
# Create a DataFrame for submission using the best model
submission = pd.DataFrame({
    'Ratings': test_data['Ratings'],
    'RAM': test_data['RAM'],
    'ROM': test_data['ROM'],
    'Battery_Power': test_data['Battery_Power'],
    'Price': test_predictions
})
```

```python
# Save the submission to a CSV file
submission.to_csv('mobile_price_submission.csv', index=False)
print("Predictions saved to 'mobile_price_submission.csv'.")
```
