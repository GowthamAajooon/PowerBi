```python
#Exploratory Data Analysis 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import missingno as msno
```

```python
df = pd.read_csv("1TSLA.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year from the 'Date' column and create a new column 'Year'
df['Year'] = df['Date'].dt.year

# Create a histogram of the count of 'Open' prices based on year
plt.figure(figsize=(10, 6))
df['Year'].value_counts().sort_index().plot(kind='bar')

# Set plot title and labels
plt.title('Count of Open Prices Based on Year')
plt.xlabel('Year')
plt.ylabel('Count of Open Prices')

# Show the plot
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.plot(df['Date'], df['Close'], label='Close')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Tesla Stock Prices Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Date'], df['Volume'])

plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Tesla Stock Volume Traded Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Close'], bins=5, edgecolor='black')

plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.title('Distribution of Tesla Close Prices')
plt.grid(True)
plt.show()
```

```python
# Plot the box plot
plt.figure(figsize=(10, 6))
plt.boxplot([df['Open'], df['High'], df['Low'], df['Close']], labels=['Open', 'High', 'Low', 'Close'])

plt.ylabel('Price')
plt.title('Box Plot of Tesla Stock Prices')
plt.grid(True)
plt.show()
```

```python
# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Open'], df['Close'])

plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.title('Scatter Plot of Open vs Close Prices')
plt.grid(True)
plt.show()
```

```python
data = pd.DataFrame({
    'date_column': ['2010-06-29', '2010-07-01', '2010-07-05'],
    'numeric_column1': [1.2, 2.4, 3.5],
    'numeric_column2': [4.5, 5.6, 6.7]
})

# Identify non-numeric columns
non_numeric_columns = data.select_dtypes(include=['object', 'datetime']).columns
print("Non-numeric columns:", non_numeric_columns)

# Option A: Convert non-numeric columns (example for date columns)
if 'date_column' in non_numeric_columns:
    data['date_column'] = pd.to_datetime(data['date_column'])
    data['date_column'] = data['date_column'].map(pd.Timestamp.timestamp)

# Drop any remaining non-numeric columns
data_numeric = data.select_dtypes(include=[float, int])
                                                                                                                                                                                 
# Calculate the correlation
correlation_matrix = data_numeric.corr()
print(correlation_matrix)
```
