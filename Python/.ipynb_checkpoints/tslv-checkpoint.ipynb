```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Histogram for a numerical column
plt.figure(figsize=(10, 6))
sns.histplot(data['High'], bins=30)
plt.title('Histogram of Numerical Column')
plt.show()

```

```python
# Bar plot for a categorical column
plt.figure(figsize=(10, 6))
sns.countplot(data['Low'])
plt.title('Bar Plot of Categorical Column')
plt.show()

```

```python
# Scatter plot for two numerical columns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='High', y='Low', data=data)
plt.title('Scatter Plot of Numerical Columns')
plt.show()
```
